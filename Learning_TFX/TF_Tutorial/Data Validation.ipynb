```python
import tensorflow_data_validation as tfvd

stats = tfvd.generate_statistics_from_csv(
    data_location = '../../Datasets/winequality-white.csv',
    delimiter=';'
)
```

```python
tfvd.visualize_statistics(stats)
```

```python
schema = tfvd.infer_schema(stats)
schema
```

```python
tfvd.display_schema(schema)
```

```python
tfvd.get_feature(schema, 'density').presence.min_fraction = 0.7
```

```python
tfvd.display_schema(schema)
```

```python
anomalies = tfvd.validate_statistics(statistics=stats, schema=schema)
anomalies
```

```python
tfvd.display_anomalies(anomalies)
```
